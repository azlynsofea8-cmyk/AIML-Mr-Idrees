```python
import matplotlib.pyplot as plt
import pandas as pd

# Example dataset
data2 = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, 35, 200, 40],  # 200 is an outlier
}

df = pd.DataFrame(data2)
df
```

```python
plt.boxplot(df['Age'])
```

### Z SCORE - how many standard deviations a data point is away form the mean

```python
# std deviation for the age

# all points
# now we have a formula too
# All data points will have a z score
# if the z score > 3 or < -3 you can treat them as outliers and drop them 

```

```python
pip install scipy
```

```python
from scipy import stats

z_scores = stats.zscore(df['Age'])
type(z_scores)
```

```python
threshold = 1.5

#  -3 < z < 3

new_x = df[  (-threshold< z_scores) & (z_scores < threshold)  ]
new_x
```

```python
plt.boxplot(new_x['Age'])
```
